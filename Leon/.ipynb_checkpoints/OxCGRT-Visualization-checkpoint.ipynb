```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
% ls Datasets/OxCGRT/
```

```python
df = pd.read_csv("~/SageMaker/Datasets/OxCGRT/OxCGRT.csv")
df.Date = pd.to_datetime(df.Date, format="%Y%m%d")
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains("Unnamed") & ~df.columns.str.contains("_Notes")]
```

```python
def diff_for_cases(data):
    diff = data[["ConfirmedCases", "ConfirmedDeaths"]].diff()
    
    data["NewConfirmedCases"] = diff["ConfirmedCases"]
    data["NewConfirmedDeaths"] = diff["ConfirmedDeaths"]
    
    return data

df = df.groupby(["CountryName"]).apply(diff_for_cases).reset_index(drop=True)
df.to_csv("~/SageMaker/Datasets/OxCGRT/OxCGRT-Processed.csv", index=False)
```

```python
df["ConfirmedCases"].fillna(0, inplace=True)
df["ConfirmedDeaths"].fillna(0, inplace=True)
df["StringencyIndexForDisplay"].fillna(0, inplace=True)
```

```python
df.head()
```

```python
def plot_log_cases_for_country(df, country_name):
    df_country = df[df.CountryName == country_name]
    df_country["days"] = (df_country.Date - df_country.Date.min()).dt.days

    plt.plot(df_country.days, np.log(df_country.ConfirmedCases + 1))
    plt.plot(df_country.days, np.log(df_country.ConfirmedDeaths + 1))
    plt.plot(df_country.days, df_country.StringencyIndexForDisplay)
    
    plt.show()

plot_log_cases_for_country(df, "Ireland")
```

```python
def plot_increase_per_country(df, country_name):
    df_country = df[df.CountryName == country_name]

    df_country["days"] = (df_country.Date - df_country.Date.min()).dt.days

    df_country_diff = df_country[["days", "ConfirmedCases", "ConfirmedDeaths"]].set_index('days').diff()
    df_country_diff.fillna(0, inplace=True)
    
    max_new_cases = df_country_diff.ConfirmedCases.max()

    plt.plot(df_country.days, df_country_diff.ConfirmedCases * 100 / max_new_cases)
    plt.plot(df_country.days, df_country_diff.ConfirmedDeaths * 100 / max_new_cases)
    plt.plot(df_country.days, df_country.StringencyIndexForDisplay)
    
    plt.show()
    
plot_increase_per_country(df, "Ireland")
```

```python
plot_increase_per_country(df, "Croatia")
```

```python
plot_increase_per_country(df, "China")
```

```python
plot_increase_per_country(df, "Italy")
```

```python
def plot_number_of_days_till_indexed_measure(data, index, index_general=None):
    from collections import defaultdict
    
    def f(data):
        data = data[data.ConfirmedCases > 0]
        day1 = data.Date.min()
        
        data["days_since_first"] = (data.Date - day1).dt.days
        
        if index_general is None:
            cond = data[index] > 0
        else:
            cond = (data[index] > 0) & (data[index_general] > 0)
            
        if cond.any():
            measure_day = data[cond].iloc[0]
            days_until_measure = measure_day.days_since_first
        else:
            days_until_measure = np.nan
        
        return pd.DataFrame([[days_until_measure]], columns=["DaysUntilMeasure"])

        
    result = data.groupby(["CountryCode", "CountryName"]).apply(f).reset_index()
    
    plt.figure(figsize=(10,20))
    plt.scatter(np.zeros(len(result)), result.DaysUntilMeasure)
    
    position_text_dict = defaultdict(list)
    for _, row in result.iterrows():
        position_text_dict[row.DaysUntilMeasure].append(row.CountryCode)
        
    for position in position_text_dict:
        text = ", ".join(position_text_dict[position])
        plt.annotate(text, (0, position), xytext=(0.0005, position), ha='left')
    
    plt.title(f"Days between first case and {index}")
    plt.show()
    
    print("Countries that never implemented the measure:", list(result[pd.isna(result.DaysUntilMeasure)].CountryName))
```

```python
plot_number_of_days_till_indexed_measure(df, "S1_School closing", index_general="S1_IsGeneral")
```

```python
plot_number_of_days_till_indexed_measure(df, "S2_Workplace closing", index_general="S2_IsGeneral")
```
