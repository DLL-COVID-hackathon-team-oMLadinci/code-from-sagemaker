```python
import pandas as pd
```

```python
df_feat_format = pd.read_csv("~/SageMaker/Datasets/features_format.csv")
df_feat_format["date"] = pd.to_datetime(df_feat_format["date"])

df_feat_format.head()
```

```python
df_feat_format = df_feat_format.loc[:, ~df_feat_format.columns.str.contains("Unnamed")]
df_feat_format.head()
```

```python
df_crgt = pd.read_csv("~/SageMaker/Datasets/OxCGRT/OxCGRT-Processed.csv")
df_crgt["Date"] = pd.to_datetime(df_crgt["Date"])

del df_crgt["ConfirmedCases"]
del df_crgt["ConfirmedDeaths"]
del df_crgt["NewConfirmedCases"]
del df_crgt["NewConfirmedDeaths"]
del df_crgt["CountryName"]

df_crgt.head()
```

```python
joined = df_feat_format.join(df_crgt.set_index(["CountryCode", "Date"]), on=["country_code", "date"], how="left").reset_index(drop=True)
joined = joined.loc[:, ~joined.columns.str.contains("Unnamed")]
joined.head()
```

```python
joined.to_csv("~/SageMaker/Datasets/features_format_cgrt_AI.csv", index=False)
```

```python
missing_countries = set(df_crgt.CountryCode).difference(set(joined.country_code))

df = pd.read_csv("~/SageMaker/Datasets/OxCGRT/OxCGRT-Processed.csv")
df[df.CountryCode.isin(missing_countries)].CountryName.unique()
```
