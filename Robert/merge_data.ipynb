```python
import pandas as pd
```

```python
cases = pd.read_csv("../cases.csv", parse_dates=True)
```

```python
deaths = pd.read_csv("../deaths.csv", parse_dates=True)
```

```python
cases.sort_values("Date")
```

```python
deaths = deaths.sort_values(["Date", "Country"])
cases = cases.sort_values(["Date", "Country"])
```

```python
deaths.head(5)
```

```python
cases.head(5)
```

```python
deaths = deaths.reset_index(drop=True)
cases = cases.reset_index(drop=True)
```

```python
deaths = deaths.drop(columns=["Country", "Date"])
```

```python
df = deaths.join(cases)
```

```python
df.to_csv("../final.csv", index=False)
```
