```python
import pandas as pd
```

## Cumulative

__Confirmed cases__

```python
cf = pd.read_csv("../Datasets/John_Hopkins/Raw/FIXED_time_series_covid19_confirmed_global.csv")
```

```python
cf["Country/Region"].unique()
```

```python
cf = cf.drop(columns=['Province/State', 'Lat', 'Long'])
cf = cf.groupby('Country/Region').sum() 
cf.head(5)
```

```python
cf = cf.transpose()
```

```python
cf.index.names = ['Date']
cf.columns.names = [""]
```

```python
cf.head(20)
```

```python
cf.to_csv("../Datasets/John_Hopkins/Clean/confirmed_cumulative.csv")
```

__Deaths and Recovered__

```python
df = pd.read_csv("../Datasets/John_Hopkins/Raw/FIXED_time_series_covid19_deaths_global.csv")
rf = pd.read_csv("../Datasets/John_Hopkins/Raw/FIXED_time_series_covid19_recovered_global.csv")
```

```python
df = df.drop(columns=['Province/State', 'Lat', 'Long'])
df = df.groupby('Country/Region').sum() 
df = df.transpose()
df.index.names = ['Date']
df.columns.names = [""]
```

```python
rf = rf.drop(columns=['Province/State', 'Lat', 'Long'])
rf = rf.groupby('Country/Region').sum() 
rf = rf.transpose()
rf.index.names = ['Date']
rf.columns.names = [""]
```

```python
df.head(10)
```

```python
rf.head(10)
```

```python
rf['Italy']
```

```python
df.to_csv("../Datasets/John_Hopkins/Clean/deaths_cumulative.csv")
```

```python
rf.to_csv("../Datasets/John_Hopkins/Clean/recovered_cumulative.csv")
```

## Daily Increase

```python
cff = cf.diff()
cff.fillna(0, inplace=True)
```

```python
cff.head(5)
```

```python
dff = df.diff()
dff.fillna(0, inplace=True)
rff = rf.diff()
rff.fillna(0, inplace=True)
```

```python
rff.to_csv("../Datasets/John_Hopkins/Clean/recovered_daily.csv")
dff.to_csv("../Datasets/John_Hopkins/Clean/deaths_daily.csv")
cff.to_csv("../Datasets/John_Hopkins/Clean/confirmed_daily.csv")
```
