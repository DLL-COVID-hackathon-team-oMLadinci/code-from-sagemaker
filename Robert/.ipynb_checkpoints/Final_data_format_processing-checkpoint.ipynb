```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Converting to Features Format

```python
cf = pd.read_csv("../Datasets/John_Hopkins/Clean/confirmed_daily.csv", index_col=0)
df = pd.read_csv("../Datasets/John_Hopkins/Clean/deaths_daily.csv", index_col=0)
rf = pd.read_csv("../Datasets/John_Hopkins/Clean/recovered_daily.csv", index_col=0)
```

```python
cf = cf.reset_index()
cf = cf.melt(id_vars="Date")
cf.columns = ["date", "country", "confirmed"]
```

```python
cf.head(10)
```

```python
df = df.reset_index()
df = df.melt(id_vars="Date")
df.columns = ["date", "country", "deaths"]
```

```python
rf = rf.reset_index()
rf = rf.melt(id_vars="Date")
rf.columns = ["date", "country", "recovered"]
```

```python
cf["recovered"] = rf["recovered"]
cf["deaths"] = df["deaths"]
```

```python
cf.to_csv("../Datasets/features_format.csv")
```

```python
df = cf
```

### Adding Country Codes

```python
c_dict = pd.read_csv("../country_dict.csv")
```

```python
c_dict["country"] = c_dict["country"].str.replace("_", " ")
c_dict = c_dict.set_index("country")
c_dict.head(5)
```

```python
code_dict = c_dict.to_dict()['code']
```

```python
dict_entries = set(code_dict.keys())
df_entries = set(df["country"].unique())
```

```python
df_entries.difference(dict_entries)
```

```python
dict_entries.difference(df_entries)
```

```python
correction_dict = {
    "Brunei": "Brunei Darussalam",
    "Burma": "Myanmar",
    "Cabo Verde": "Cape Verde",
    "Congo (Brazzaville)": "Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo",
    "Cote d'Ivoire": "Cote dIvoire",
    "Guinea-Bissau": "Guinea Bissau",
    "Korea, South": "South Korea",
    "Taiwan*": "Taiwan",
    "Tanzania": "United Republic of Tanzania",
    "Timor-Leste": "Timor Leste",
    "US": "United States of America",
    "West Bank and Gaza": "Palestine",
    "Turks and Caicos Islands": "Turks and Caicos islands",
    "Curacao": "Cura√ßao"
}
```

```python
df["country"] = df["country"].apply(lambda x: correction_dict[x] if x in correction_dict.keys() else x)
```

```python
dict_entries = set(code_dict.keys())
df_entries = set(df["country"].unique())
df_entries.difference(dict_entries)
```

```python
dict_entries.difference(df_entries)
```

```python
df["country_code"] = df["country"].map(code_dict)
```

```python
df = df.dropna()
```
