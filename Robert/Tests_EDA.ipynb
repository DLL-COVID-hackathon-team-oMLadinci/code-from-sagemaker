```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Datasets/Wikipedia/tests.csv")
```

```python
df = df.sort_values("%")
```

```python
df.columns = ['country', "date", "n_tests", "n_positive", "pct_positive", "test_by_million", "pos_by_million"]
```

```python
df.head(5)
```

```python
df['test_by_million'] = df['test_by_million'].str.replace(",", "")
df['n_tests'] = df['n_tests'].str.replace(",", "")
df['n_positive'] = df['n_positive'].str.replace(",", "")
```

```python
df['test_by_million']=df['test_by_million'].astype('int32')
df['n_tests']=df['n_tests'].astype('int32')
df['n_positive']=df['n_positive'].astype('int32')
```

```python
df.head(10)
```

```python
objects = df.country
y_pos = np.arange(len(objects))

percent_pos = df.pct_positive
test_mill = df.test_by_million

fig, axs = plt.subplots(2, figsize=(30,20), sharex=True)
axs[0].bar(y_pos, percent_pos, align='center', alpha=0.5)
axs[1].bar(y_pos, test_mill, align='center', alpha=0.5)

axs[0].set_xticks(y_pos)
axs[0].set_xticklabels(objects, rotation=90)
axs[1].set_xticks(y_pos)
axs[1].set_xticklabels(objects, rotation=90)
#plt.xticks(rotation='vertical')

plt.show()
```
